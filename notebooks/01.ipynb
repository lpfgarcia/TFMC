## Loading a dataset

```python
from datasets import load_dataset
```

```python
raw = load_dataset('code_search_net', 'python') 
```

```python
raw
```

```python
raw['train'].features
```

```python
print(raw['train'][99]['whole_func_string'])
```

```python
def get_training_corpus(raw):
    dataset = raw['train']
    for idx in range(0, len(dataset), 1000):
        samples = dataset[idx : idx + 1000]
        yield samples['whole_func_string']
```

```python
trainig_corpus = get_training_corpus(raw)
```

## Retraining the tokenizer

```python
new_tokenizer = tokenizer.train_new_from_iterator(trainig_corpus, 52000)
```

```python
new_tokenizer.all_special_tokens
```

```python
new_tokenizer.all_special_ids
```

```python
len(new_tokenizer.get_vocab())
```

```python
new_tokenizer.get_vocab()
```

```python
new_tokenizer.get_vocab()['def']
```

## Testing the tokenizers (old and new)

```python
examples = [
    ['''def add_numbers(a, b):
    """Add the two numbers `a` and `b`."""
    return a + b''', 
     '''class LinearLayer():
    def __init__(self, input_size, output_size):
        self.weight = torch.randn(input_size, output_size)
        self.bias = torch.zeros(output_size)

    def __call__(self, x):
        return x @ self.weights + self.bias
    ''']]
```

```python
print(tokenizer.tokenize(examples))
```

```python
print(new_tokenizer.tokenize(examples))
```
